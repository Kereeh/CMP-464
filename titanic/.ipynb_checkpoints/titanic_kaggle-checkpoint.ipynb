687 out of 891 data points are missing from the Cabin column, it would be best to drop this value
177 out of 891 data points are missing from Age value - how should I replace this data? impute with mean or median?

```python
#dropping cabin as irrelevant value
train_set = train_set.drop(columns = ['Cabin'])
```

```python
#filling NA Age values with median age 
train_set['Age'] = train_set['Age'].fillna(train_set['Age'].median())
```

```python
#replacing 2 missing embarked values with port most boarded by people - S
train_set['Embarked'] = train_set['Embarked'].fillna('S')
```

```python
#extracting honorifics from the name variable
train_set['Title'] = train_set['Name'].str.extract(' ([A-za-z]+)\.', expand=False)
```

```python
#check if missing values is adjusted in the training data
train_set.isnull().sum()
```

```python
#creating categorical variables and dropping repeating variables
#class, embarked, and sex

new_train_set = pd.get_dummies(train_set, columns = ['Pclass', 'Embarked', 'Sex'])

#repetition of sex_male
new_train_set.drop('Sex_female', axis=1, inplace = True)

#name is irrelevant since title is extracted
new_train_set.drop('Name', axis=1, inplace = True)

#ticket is irrelevant
new_train_set.drop('Ticket', axis=1, inplace = True)

new_train_set.head()
```

```python
#sibling, spouse, parents, and children simply identify whether or not you traveled alone or with famile
#let's convert it to a binary category of 'alone or nah'

new_train_set['Alone'] = np.where((new_train_set['SibSp'] + new_train_set['Parch'])>0, 0, 1)
new_train_set.drop('Parch', axis =1, inplace = True)
new_train_set.drop('SibSp', axis =1, inplace = True)
```

```python
new_train_set.drop('Title', axis =1, inplace = True)
train_set = new_train_set

train_set.head()
```

```python
from sklearn.model_selection import cross_validate

scoring = {'accuracy' : 'accuracy', 'log_loss': 'neg_log_loss', 'auc': 'roc_auc'}

model_cv = LogisticRegression()

results = cross_validate(model_cv, X, y, cv = 5, scoring = list(scoring.values()), return_train_score = False)

print('5-fold cross-validation results:')
for sc in range(len(scoring)):
    print(model_cv.__class__.__name__+" average %s: %.3f (+/-%.3f)" % (list(scoring.keys())[sc], -results['test_%s' % list(scoring.values())[sc]].mean()
                               if list(scoring.values())[sc]=='neg_log_loss' 
                               else results['test_%s' % list(scoring.values())[sc]].mean(), 
                               results['test_%s' % list(scoring.values())[sc]].std()))
```
